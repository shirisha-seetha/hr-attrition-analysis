```python
#=================================
# HR Attrition Analysis
#==================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

sns.set_style('whitegrid')

# Load cleaned data
df = pd.read_csv('C:\\Users\\SSS\\OneDrive\\Desktop\\hr-attrition-analysis\\data\\processed\\hr_attriton_cleaned.csv')
print(f"Data Loaded — Shape: {df.shape}")
```

```python
# ----------------------------------------
# 1. OVERALL ATTRITION RATE
# ----------------------------------------
total = len(df)
attrition = df['Attrition_flag'].sum()
rate = df['Attrition_flag'].mean() * 100

print("=" * 45)
print("        HR ATTRITION — KEY INSIGHTS")
print("=" * 45)
print(f"Total Employees      : {total}")
print(f"Employees Left       : {attrition}")
print(f"Overall Attrition %  : {rate:.2f}%")
print("=" * 45)
```

```python
# ----------------------------------------
# 2. INSIGHT 1 — DEPARTMENT
# ----------------------------------------
dept = df.groupby('Department')['Attrition_flag'].mean() * 100
highest_dept = dept.idxmax()
highest_rate = dept.max()

print(f"\n🔴 INSIGHT 1 — DEPARTMENT")
print(f"Highest attrition is in '{highest_dept}' department")
print(f"Attrition Rate: {highest_rate:.1f}%")
print(dept.sort_values(ascending=False).to_string())
```

```python
# ----------------------------------------
# 3. INSIGHT 2 — AGE GROUP
# ----------------------------------------
df['Age_Group'] = pd.cut(df['Age'],
                          bins=[18, 25, 35, 45, 60],
                          labels=['18-25', '26-35', '36-45', '46-60'])

age = df.groupby('Age_Group')['Attrition_flag'].mean() * 100
highest_age = age.idxmax()
highest_age_rate = age.max()

print(f"\n🔴 INSIGHT 2 — AGE GROUP")
print(f"Highest attrition in age group: '{highest_age}'")
print(f"Attrition Rate: {highest_age_rate:.1f}%")
print(age.sort_values(ascending=False).to_string())
```

```python
# ----------------------------------------
# 4. INSIGHT 3 — OVERTIME
# ----------------------------------------
ot = df.groupby('OverTime')['Attrition_flag'].mean() * 100

ot_yes = ot.get('Yes', 0)
ot_no = ot.get('No', 0)
diff = ot_yes - ot_no

print(f"\n🔴 INSIGHT 3 — OVERTIME IMPACT")
print(f"Attrition with Overtime    : {ot_yes:.1f}%")
print(f"Attrition without Overtime : {ot_no:.1f}%")
print(f"Employees doing overtime are {diff:.1f}% more likely to leave!")
```

```python
# ----------------------------------------
# 5. INSIGHT 4 — MONTHLY INCOME
# ----------------------------------------
income_yes = df[df['Attrition'] == 'Yes']['MonthlyIncome'].mean()
income_no = df[df['Attrition'] == 'No']['MonthlyIncome'].mean()
diff_income = income_no - income_yes

print(f"\n🔴 INSIGHT 4 — MONTHLY INCOME")
print(f"Avg Income (Left)     : ${income_yes:,.0f}")
print(f"Avg Income (Stayed)   : ${income_no:,.0f}")
print(f"Employees who left earned ${diff_income:,.0f} less on average")
```

```python
# ----------------------------------------
# 6. INSIGHT 5 — JOB SATISFACTION
# ----------------------------------------
js = df.groupby('JobSatisfaction')['Attrition_flag'].mean() * 100
lowest_js = js.idxmin()
highest_js_rate = js.max()

print(f"\n🔴 INSIGHT 5 — JOB SATISFACTION")
print(f"Highest attrition at satisfaction level: {lowest_js}")
print(f"Attrition Rate: {highest_js_rate:.1f}%")
print(js.sort_values(ascending=False).to_string())
```

```python
# ----------------------------------------
# 7. INSIGHT 6 — YEARS AT COMPANY
# ----------------------------------------
early = df[df['YearsAtCompany'] <= 2]['Attrition_flag'].mean() * 100
mid = df[(df['YearsAtCompany'] > 2) & (df['YearsAtCompany'] <= 5)]['Attrition_flag'].mean() * 100
senior = df[df['YearsAtCompany'] > 5]['Attrition_flag'].mean() * 100

print(f"\n🔴 INSIGHT 6 — YEARS AT COMPANY")
print(f"0-2 Years (Early)    : {early:.1f}% attrition")
print(f"3-5 Years (Mid)      : {mid:.1f}% attrition")
print(f"5+ Years (Senior)    : {senior:.1f}% attrition")
print(f"Early employees are most likely to leave!")
```

```python
# ----------------------------------------
# 8. FINAL SUMMARY — BUSINESS RECOMMENDATIONS
# ----------------------------------------
print("\n" + "=" * 45)
print("       BUSINESS RECOMMENDATIONS")
print("=" * 45)

recommendations = {
    "1": "Focus retention efforts on Sales department — highest attrition",
    "2": "Young employees (18-25) need better career growth paths",
    "3": "Reduce overtime — it significantly increases attrition risk",
    "4": "Review salary structure — low earners leave more often",
    "5": "Improve onboarding — most exits happen in first 2 years",
    "6": "Boost job satisfaction through engagement programs"
}

for key, value in recommendations.items():
    print(f"\n✅ Recommendation {key}:")
    print(f"   {value}")

print("\n" + "=" * 45)
```

```python
# ----------------------------------------
# 9. SAVE SUMMARY AS TEXT FILE
# ----------------------------------------
summary = f"""
HR ATTRITION ANALYSIS — INSIGHTS SUMMARY
==========================================
Total Employees      : {total}
Employees Left       : {attrition}
Overall Attrition %  : {rate:.2f}%

KEY INSIGHTS
------------
1. {highest_dept} dept has highest attrition at {highest_rate:.1f}%
2. Age group {highest_age} leaves the most at {highest_age_rate:.1f}%
3. Overtime employees are {diff:.1f}% more likely to leave
4. Employees who left earned ${diff_income:,.0f} less on average
5. Early employees (0-2 yrs) have {early:.1f}% attrition rate

RECOMMENDATIONS
---------------
1. Focus retention on Sales department
2. Create growth paths for young employees
3. Reduce or compensate overtime fairly
4. Review and improve salary structure
5. Strengthen onboarding program
"""

with open(r'C:\Users\SSS\OneDrive\Desktop\hr-attrition-analysis\reports\insights_summary.txt', 'w') as f:
    f.write(summary)

print("✅ Summary saved to /reports/insights_summary.txt")
```
